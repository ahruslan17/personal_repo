```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных о цене Биткоина за 2022 год
btc = yf.download("BTC-USD", start="2022-01-01", end="2023-01-01")

# Создание графика цен закрытия
plt.figure(figsize=(8, 4))
plt.plot(btc['Close'])
plt.title("Цена закрытия Биткоина за 2022 год")
plt.xlabel("Дата")
plt.ylabel("Цена закрытия")
plt.show()

# Вычисление ряда доходности цен
btc_returns = btc['Close'].pct_change().dropna()

# Создание графика ряда доходности цен
plt.figure(figsize=(8, 4))
plt.plot(btc_returns)
plt.title("Ряд доходности цен Биткоина за 2022 год")
plt.xlabel("Дата")
plt.ylabel("Доходность")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных о цене Биткоина за 2022 год
btc = yf.download("BTC-USD", start="2022-01-01", end="2023-01-01")

# Создание графика цен закрытия
plt.figure(figsize=(8, 4))
plt.plot(btc['Close'])
plt.title("Цена закрытия Биткоина за 2022 год")
plt.xlabel("Дата")
plt.ylabel("Цена закрытия")
plt.show()

# Вычисление ряда доходности цен
btc_pct_change = btc['Close'].pct_change()

# Применение экспоненциального сглаживания
alpha = 0.5
btc_ewma_returns = btc_pct_change.ewm(alpha=alpha).mean().dropna()

# Создание графика сглаженного ряда доходности цен
plt.figure(figsize=(8, 4))
plt.plot(btc_ewma_returns)
plt.title("Сглаженный ряд доходности цен Биткоина за 2022 год")
plt.xlabel("Дата")
plt.ylabel("Сглаженная доходность")
plt.show()
```
