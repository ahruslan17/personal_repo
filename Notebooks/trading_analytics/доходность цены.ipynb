```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#       2022 
btc = yf.download("BTC-USD", start="2022-01-01", end="2023-01-01")

#    
plt.figure(figsize=(8, 4))
plt.plot(btc['Close'])
plt.title("    2022 ")
plt.xlabel("")
plt.ylabel(" ")
plt.show()

#    
btc_returns = btc['Close'].pct_change().dropna()

#     
plt.figure(figsize=(8, 4))
plt.plot(btc_returns)
plt.title("     2022 ")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#       2022 
btc = yf.download("BTC-USD", start="2022-01-01", end="2023-01-01")

#    
plt.figure(figsize=(8, 4))
plt.plot(btc['Close'])
plt.title("    2022 ")
plt.xlabel("")
plt.ylabel(" ")
plt.show()

#    
btc_pct_change = btc['Close'].pct_change()

#   
alpha = 0.5
btc_ewma_returns = btc_pct_change.ewm(alpha=alpha).mean().dropna()

#      
plt.figure(figsize=(8, 4))
plt.plot(btc_ewma_returns)
plt.title("      2022 ")
plt.xlabel("")
plt.ylabel(" ")
plt.show()
```
