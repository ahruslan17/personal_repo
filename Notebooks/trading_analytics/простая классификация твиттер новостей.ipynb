```python
def classification(news):
    count = []
#     print('Count: ', count)
    for i in indexing:
        count_1 = 0
        pool = []
        for j in df[i].tolist():
            str1 = str(j)
            if str1 != 'nan':
                pool.append('[\W]' + str1 + '[\W]')
                
        for t in pool:     
            result = re.findall(t, news)
            count_1 += len(result)
#             print(result)
        count.append(count_1)

#     print(count)
#     print(indexing[count.index(max(count))]) 
    if max(count) == 0:
        return [count, 'None']
    return [count, indexing[count.index(max(count))]]
```

```python
def classification(news):
    count = []
    red_zone = ['[\W]lists[\W]', '[\W]partners[\W]','[\W]growth[\W]','[\W]event[\W]', '[\W]is available[\W]', '[\W]events[\W]']
    yellow_zone = ['[\W]rewards[\W]', '[\W]upgrade[\W]', '[\W]update[\W]','[\W]release[\W]','[\W]updates[\W]','[\W]upgrades[\W]', '[\W]chance to win[\W]', '[\W]hackathon[\W]', '[\W]is listed[\W]', '[\W]integrations[\W]','[\W]big update[\W]', '[\W]big update[\W]']
    blue_zone = ['[\W]voting[\W]','[\W]integrated[\W]']
#     print('Count: ', count)
    for i in indexing:
        count_1 = 0
        pool = []
        for j in df[i].tolist():
            str1 = str(j)
            if str1 != 'nan':
                pool.append('[\W]' + str1 + '[\W]')
                
        for t in pool:  
#             print(t)
            weight = 1
            result = re.findall(t, ' ' + news.lower() + ' ')

            if t in red_zone:
                weight = 0.3
                count_1 += len(result)*weight
                
            elif t in yellow_zone:
                weight = 0.5
#                 result = re.findall(t, news)
                count_1 += len(result)*weight

            elif t in blue_zone:
                weight = 0.7
#                 result = re.findall(t, news)
                count_1 += len(result)*weight
    
            else:
                count_1 += len(result)*0.9
                
#             print(result)
        count.append(count_1)

#     print(count)
#     print(indexing[count.index(max(count))]) 

    out = []
    for i in range(len(count)):
        if count[i] > 0 and count[i] <= 0.3:
            out.append(str(round(count[i]/sum(count), 3) * 100 / 5) + ' % - ' + str(indexing[i]))
            
        if count[i] > 0.3 and count[i] <= 0.5:
            out.append(str(round(count[i]/sum(count), 3) * 100 / 3) + ' % - ' + str(indexing[i]))
        
        if count[i] > 0.5 and count[i] < 0.7:
            out.append(str(round(count[i]/sum(count), 3) * 100 / 2) + ' % - ' + str(indexing[i]))
            
        if count[i] >= 0.7 and count[i] <= 0.8:
            out.append(str(round(count[i]/sum(count), 3) * 100 * 0.85) + ' % - ' + str(indexing[i]))
            
        if count[i] > 0.8 and count[i] < 1.1:
            out.append(str(round(count[i]/sum(count), 3) * 100 * 0.9) + ' % - ' + str(indexing[i]))
        
        if count[i] >= 1.1:
            out.append(str(round(count[i]/sum(count), 3) * 100) + ' % - ' + str(indexing[i]))
    if max(count) == 0:
        return 'None'
    return out
```
